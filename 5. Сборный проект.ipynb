### Обработка пропусков

```python
data_isna = pd.DataFrame(data = data.isna().sum(), columns =['isna'])
data_isna['percent'] = data.isna().mean()*100  
print(data_isna)      
print('Максимальный процент пропусков:', format(data_isna['percent'].max(), '.2f'), '%')
print('В', len(data_isna.query('percent!=0')), 'столбцах есть пропуски в данных')
```

```python
data=data.query('name.isna()==False')
data=data.query('year_of_release.isna()==False')
```

```python
print("Столбец 'user_score' имеет",  data.loc[(data['user_score'] == 'tbd'), 'user_score'].count(), 
      "или", format(data.loc[(data['user_score'] == 'tbd'), 'user_score'].count()/data.shape[0]*100, '.2f'),
      '%', "строк со значением TBD")
print("Столбец 'rating' имеет",  data.loc[(data['rating'] == 'K-A'), 'rating'].count(), 
      "или", format(data.loc[(data['rating'] == 'K-A'), 'rating'].count()/data.shape[0]*100, '.2f'), 
      '%', "строк со значением K-A")
data.loc[(data['user_score'] == 'tbd'), 'user_score'] = float('nan')
data.loc[(data['rating'] == 'K-A'), 'rating'] = 'E'
data['rating'] = data['rating'].fillna('UNKNOWN');
```

**Причины пропусков:**  

*  Техничесие фактор, например проблема при загрузке или выгрузке данных;  
*  Человеческий фактор, например проблема невнимательсности при заполнении данных.  

**Обоснование заполнения пропусков:**

1.   *Столбец:* "name" и "genre" имели 2 или 0,01% одинаковых пропусков. Были удалены.
    *  *Обоснование*: заполнить данную информацию не представляется возможным, а также пропущенные данные составляли менее 5% от всей информации. 


2.   *Столбец*: "year_of_release" имел 269 или 1,6% пропусков. Были удалены.
    *  *Обоснование*: заполнить данную информацию не представляется возможным, а также пропущенные данные составляли менее 5% от всей информации.


3.   *Столбец*: "user_score" имеет 2376 или 14.5% строк с аббревиатурой "tbd". Были заменены на значение NaN.  
    *  *Обоснование*: *TBD (to be determined)* означает, что рейтинг пока не определен,возможно, игра еще не успела получить отзывы геймеров. данная информация не может быть удалена или существенно изменена, т.к. пропущенные данные составляют более 5% от всех данных, иначе это может привести к искажению результатов исследования.  


4.   *Столбец*: "rating" имеет 3 или 0.02 % строк с аббревиатурой "K-A". Были заменены на значение "E". 
    *  *Обоснование*: "K-A" - это старое название рейтинга "E".  


5.   *Столбец*: "rating" имеет 6766 или 40,5% пропусков. Были заменены на значение "UNKNOWN". 
    *  *Обоснование*: данная информация не может быть не может быть удалена или существенно изменена, т.к. пропущенные данные составляют более 5% от всех данных, иначе это может привести к искажению результатов исследования.

## Глава №3: Исследовательский анализ

### Продажи игр по годам

```python
plt.figure(figsize=(12, 4))
plt.xlabel('Годы')
plt.ylabel('Количество продаж')
plt.title('Продажи по годам')
data.groupby('year_of_release')['name'].count().plot(label='count', marker ='o');
data.groupby('year_of_release')['sum_sales'].sum().plot(label='sum', marker ='o');
plt.legend(loc='upper left')
plt.grid()
data=data.query('year_of_release>=1994')
```

С начала наблюдений количество релизов не превышало порог в 100 игр в год. Однако количество релизов и суммы продаж начали существенный рост с 1994 по 2009 год. После этого наблюдается огромный спад. Он вызван множетсвом причин. Во-первых, кризис 2008 года повлиял на все отрасли и игровая индустрия не стала исключением, интересным моментом является тот факт, что спад продаж начался не сразу. В кризисное время человеку необходимо отвлечься от стресса и внешних раздражителей, поэтому на 2008 год пришелся пик продаж. Однако уже в скором времени продажи начали закономерно снижаться из-за отсутствия денег у покупателей. Это катастрофически повлияло на стоимость акций некоторых компаний. Например EA потеряла в тот момент 75% своего капитала. Во-вторых, с 2008 года начинается рост мобильного гейминга, что также негативно повлияло на продажи консольных и ПК игр. Прочими, но не менее важными факторами стоит назвать, увелечение стоимости комплектующих к компьютерам, участившиеся случаи терактов и стрельбы в школах, разделение рынка на инди и AAA проекты, "приетость" пользователя к одним и тем же шаблонам и т.д.


Наиболее важным временем в нашем исследовании следует считать промежуток с 1994 по 2015 год, т.к. на этот момент уже есть современные подходы к созданию видеоигр, которые используются до сих пор. 

### Продажи игр по платформам

```python
data_pivot = data.pivot_table(index='platform', values='sum_sales', aggfunc='sum')
list = data_pivot.sort_values(by='sum_sales', ascending=False).head(6).index.tolist()
data_platforms = data.query("platform in list or platform=='PC'").pivot_table(index='platform', columns='year_of_release', values='sum_sales', aggfunc='sum')
plt.figure(figsize = (12,4))
ax=sns.heatmap(data_platforms, cmap="BuPu");
```

```python
data_pivot_duration = data.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max']).reset_index()
data_pivot_duration.columns = ['platforms', 'first_year', 'last_year']
data_pivot_duration = data_pivot_duration.query('platforms in list')
data_pivot_duration['duration'] = data_pivot_duration['last_year']-data_pivot_duration['first_year']
data_pivot_duration = data_pivot_duration.sort_values("duration", ascending=False).reset_index()
data_pivot_duration
```

Наиболее прибыльными консолями стали Xbox360, Wii, PlayStation3, PlayStation2, PlayStation и DS. Эти консоли представлены тремя крупными компаниями Microsoft, Sony и  Nintendo. Стоит отметить, что эти компании так и остались ведущими на рынке. 
Примечательно, что модели из этого списка живут строго определенный срок, в среднем 10 лет, это можно объяснить как появлением новых конкурентов, так и устариванием "железа". Но нужно понимать, что полноценный срок жизни консоли составляет примерно 3-5 лет (самые активные продажи).  


В исследование намерено добавлен PC т.к. он является не стандартной консолью, а полноценным портативным компьютером. И хотя продажи оставляют желать лучшего, PC уникален сроком своего существования на рынке (от начала создания игр до современности).

### Выбор актуального периода 

Актуальным периодом слудет считать последние 3 года, т.к. в это время количество продаж стабилизировалось и это примерный активный срок жизни консоли. 

```python
data_actual = data.query('year_of_release>=2013')
```

### Лидеры продаж

Какие платформы лидируют по продажам, растут или падают? Выберите несколько потенциально прибыльных платформ.

```python
my_list = data_actual['platform'].unique()

plt.figure(figsize=(12, 4))

for col in my_list:
    data_actual_pivot_col = data_actual.query('platform == @col').pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum')
    print('Количество продаж на платформе', col, 'за последние 3 года изменились на:', format(int(data_actual_pivot_col['sum_sales'].tail(1))/int(data_actual_pivot_col['sum_sales'].head(1))*100-100, '.2f'), "%" )
    plt.plot(data_actual_pivot_col, marker ='o', label=col)
    
    plt.legend(loc='upper left') 
plt.xlabel('Годы')
plt.ylabel('Количество продаж')
plt.grid()
```

Новое поколение консолей (PS4, XONE) демонстрирует существенный рост продаж. Устаревшие консоли, хоть и присутствуют на рынке, но не занимают его существунную долю, и постепенно уходят с рынка (PS3, X360, DS, WII, PSP). 
Демонстрируют эффективность следующие платформы: PS4, XOne, WiiU. Также вставим в свое исследование платформы, не лишенные потенциала, такие как: 3DS, PSV, PC. 

### Boxplot

## Глава №5: Проверка гипотез

### Гипотеза №1:

Критический уровень вероятности примем за 5%. Уберем все пустые значения, для проведения подсчета.  
Нулевая гипотеза = Средние пользовательские рейтинги на платформах XOne и PC равны между собой.  
Aльтернативная гипотеза = Средние пользовательские рейтинги на платформах XOne и PC НЕ равны между собой. 

```python
alpha = 0.05
results = st.ttest_ind(data_actual.query('platform=="XOne" and user_score.isnull()==False')['user_score'], data_actual.query('platform=="PC" and user_score.isnull()==False')['user_score']) 
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
data_actual.query('platform=="XOne" and user_score.isnull()==False')['user_score'].hist(bins=20, density=True, alpha=0.3, color='b');
data_actual.query('platform=="PC" and user_score.isnull()==False')['user_score'].hist(bins=20, density=True, alpha=0.3, color='g');
```

**Вывод:** На оценки потребителей не влияет выбор платформы.

### Гипотеза №2:

Критический уровень вероятности примем за 5%. Уберем все пустые значения, для проведения подсчета.  
Нулевая гипотеза = Средние пользовательские рейтинги в жанрах Actoin и Sports равны между собой.  
Aльтернативная гипотеза = Средние пользовательские рейтинги в жанрах Actoin и Sports НЕ равны между собой.

```python
alpha = 0.05
results = st.ttest_ind(data_actual.query('genre=="Action" and user_score.isnull()==False')['user_score'], data_actual.query('genre=="Sports" and user_score.isnull()==False')['user_score']) 
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
data_actual.query('genre=="Action" and user_score.isnull()==False')['user_score'].hist(bins=20, density=True, alpha=0.3, color='b');
data_actual.query('genre=="Sports" and user_score.isnull()==False')['user_score'].hist(bins=20, density=True, alpha=0.3, color='g');
```

**Вывод:** На оценки потребителей влияет выбор жанра игры.

**Вывод главы №5:**  

Были проверены две основные гипотезы.

В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам scipy.stats.ttest_ind (array1, array2, equal_var), им мы и воспользовались в обоих случаях, т.к. нам нужно было сравнить:
   * Средние пользовательские рейтинги на платформах Xbox One и PC  
   * Средние пользовательские рейтинги в жанрах Action и Sports  

На основе данного анализа можно сделать вывод, что платформы не играют особой роли при выставлении оценок, однако жанр может повлиять на оценку игрока.
