# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

Цель исследования: Выявить взаимосвязи между ценой квартиры и ее характеристиками, представленными в датасете.

Задачи:

### Предобработка данных
#### Найдите и изучите пропущенные значения в столбцах 

```python
data_isna = pd.DataFrame(data = data.isna().sum(), columns =['isna'])          #Создаем ДФ в котором столбец №1 - количество пропусков
data_isna['percent'] = data.isna().mean()*100    #столбец №2 - процент пропусков от общего числа
print(data_isna)        #Выводим на экран таблицу, самый большой процент пропусков и количество столбцов с пропусками
print('Максимальный процент пропусков:', format(data_isna['percent'].max(), '.2f'), '%')
print('В', len(data_isna.query('percent!=0')), 'столбцах есть пропуски в данных')

#df = pd.DataFrame(round(data.isna().mean()*100,)).style.background_gradient('coolwarm')
#print(df) #Не лаботает :С Я бездарность
```

##### Заполните пропущенные значения там, где это возможно. 

```python
#data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
for loc in data['locality_name'].unique():
    median = data.loc[data['locality_name']== loc, 'ceiling_height'].median()
    data.loc[(data['ceiling_height'].isna()) & (data['locality_name']==loc), 'ceiling_height'] = median

data.loc[data['ceiling_height'].isna(), 'ceiling_height'] = data['ceiling_height'].median()
```

Как правило потолки имеют высотку, ограниченную определенными значениями, в связи с этим наиболее эффективным решением будет заменить пропуски медианным значением по населенному пункту

```python
data.loc[(data['floor']<=5), 'floors_total']  = data['floors_total'].fillna(5)
data.loc[(data['floor']<=9) & (data['floor']>5), 'floors_total']  = data['floors_total'].fillna(9)
data.loc[(data['floor']>9) & (data['floor']<=12), 'floors_total'] = data['floors_total'].fillna(12)
data.loc[(data['floor']>12) & (data['floor']<=16), 'floors_total'] = data['floors_total'].fillna(16)
data.loc[(data['floor']>16), 'floors_total'] = data['floors_total'].fillna(25)
```

В России и в Санкт-петербурге в частности пока еще преобладает застройка, которая имеет строго определеную этажность (5,9,12,16,25), в связи с этим наиболее эффективным решением будет заменить пропуски этими значениями на основе данных из столбца "floor".

```python
#data['living_area'] = data['living_area'].fillna(data['living_area'].median())
for loc in data['locality_name'].unique():
    median = data.loc[data['locality_name']== loc, 'living_area'].median()
    data.loc[(data['living_area'].isna()) & (data['locality_name']==loc), 'living_area'] = median

data.loc[data['living_area'].isna(), 'living_area'] = data['living_area'].median()
```

Как правило жилая площадь ограничена определенными значениями, в связи с этим наиболее эффективным решением будет заменить пропуски медианным значением по населенному пункту.

```python
data.loc[(data['studio']==True), 'kitchen_area'] = data['kitchen_area'].fillna(0)
#data['kitchen_area'] = data['kitchen_area'].fillna(data['kitchen_area'].median())

for loc in data['locality_name'].unique():
    median = data.loc[data['locality_name']== loc, 'kitchen_area'].median()
    data.loc[(data['kitchen_area'].isna()) & (data['locality_name']==loc), 'kitchen_area'] = median

data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = data['kitchen_area'].median()
```

В квартирах студиях как правило нет отдельной площади для кухни, в связи с этим наиболее эффективным решением будет, заполнить такие пропуски 0, а во всех остальных случаях медианным значением по населенному пункту.

```python
my_list = ['balcony', 'parks_around3000', 'ponds_around3000']
for col in my_list:
    data[col] = data[col].fillna(0)
print("Количество нулей в стоблце 'ponds_around3000' равно", data[data['ponds_around3000']==0]['ponds_around3000'].count())
print("Количество пропусков в стоблце 'ponds_nearest' равно", data['ponds_nearest'].isna().sum())
print("Количество нулей в стоблце 'parks_around3000' равно", data[data['parks_around3000']==0]['parks_around3000'].count())
print("Количество пропусков в стоблце 'parks_nearest' равно", data['parks_nearest'].isna().sum())
```

В столбце 'balcony' пропуски, скорее всего, означают, что в данном объекте недвижимости балкон отсутствует.

В столбцах 'parks_around3000' и 'ponds_around3000' пропущенные значения было решено заменить нулями. В таком случае пропущенные значения в 'parks_nearest' и 'pands_nearest' совпадут с количеством нулей в этих столбцах.

Столбец "days_exposition" не будет заполнен, т.к. новые значения могут исказить результаты исследования

Столбец "airports_nearest" не может быть заполнен никакими значениями т.к. выборка заканчивается на 80 км, а некоторые населенные пункты, например поселок Рябово находятся более чем в 100 км от аэропорта "Пулково" (~130 км).
